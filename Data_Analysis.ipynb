```python
# ðŸ“Œ Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Nice styling
plt.style.use("seaborn-v0_8")
sns.set_palette("pastel")

# ðŸ“Œ Step 2: Load the Dataset
df = pd.read_csv("Superstore (1).csv", encoding="latin1")   # adjust filename if needed

# ðŸ“Œ Step 3: Basic Exploration
print("ðŸ”¹ First 5 rows of dataset:")
display(df.head())

print("\nðŸ”¹ Dataset Info:")
print(df.info())

print("\nðŸ”¹ Summary Statistics:")
display(df.describe())

print("\nðŸ”¹ Unique Categories:")
print(df["Category"].unique())

print("\nðŸ”¹ Unique Regions:")
print(df["Region"].unique())

# ðŸ“Œ Step 4: Total Sales by Category
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
print("\nðŸ”¹ Total Sales by Category:")
print(category_sales)

category_sales.plot(kind="bar", figsize=(6,4), title="Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()

# ðŸ“Œ Step 5: Total Sales by Sub-Category
subcat_sales = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)
print("\nðŸ”¹ Total Sales by Sub-Category:")
print(subcat_sales)

subcat_sales.plot(kind="bar", figsize=(10,4), title="Total Sales by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()

# ðŸ“Œ Step 6: Total Sales by Region
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
print("\nðŸ”¹ Total Sales by Region:")
print(region_sales)

region_sales.plot(kind="pie", autopct="%1.1f%%", figsize=(6,6), title="Sales by Region")
plt.ylabel("")
plt.show()

# ðŸ“Œ Step 7: Daily Sales Trend
df["Order Date"] = pd.to_datetime(df["Order Date"])
daily_sales = df.groupby("Order Date")["Sales"].sum()

print("\nðŸ”¹ Daily Sales Trend:")
print(daily_sales.head())

daily_sales.plot(kind="line", marker="o", figsize=(12,5), title="Daily Sales Trend")
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.show()

# ðŸ“Œ Step 8: Category Sales Over Time
category_daily = df.groupby(["Order Date", "Category"])["Sales"].sum().reset_index()

plt.figure(figsize=(12,5))
sns.lineplot(data=category_daily, x="Order Date", y="Sales", hue="Category")
plt.title("Category Sales Over Time")
plt.xticks(rotation=45)
plt.show()

# ðŸ“Œ Step 9: Heatmap (Region vs Category)
pivot_table = df.pivot_table(values="Sales", index="Region", columns="Category", aggfunc="sum")

print("\nðŸ”¹ Sales Heatmap (Region vs Category):")
display(pivot_table)

plt.figure(figsize=(6,4))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="Blues")
plt.title("Sales Heatmap (Region vs Category)")
plt.show()

```
